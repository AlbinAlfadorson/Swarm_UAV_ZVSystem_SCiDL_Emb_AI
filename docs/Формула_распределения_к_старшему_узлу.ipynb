# Формула распределения сигнала в рое дронов с учётом удалённости## Основное уравнение с задержкой$$Q(t) = \sum_{i=1}^{N} \left[ \frac{W_i \cdot In_i(t - \tau_i)}{d_i^{\alpha}} \right]$$где:- $\tau_i = \dfrac{d_i}{v}$ — задержка распространения- $d_i$ — расстояние от дрона $i$ до королевы- $v$ — скорость сигнала (например, скорость света)- $\alpha$ — коэффициент затухания ($\alpha = 2$ для свободного пространства)## AMPL модель с распространением```amplset Drones;set Time;param In {i in Drones, t in Time};  # Входной сигналparam d {i in Drones};              # Расстоянияparam v;                            # Скорость сигналаparam alpha default 2;              # Затуханиеvar Q {t in Time};                  # Сигнал королевы# Задержка в тактахparam dt;                           # Длительность тактаparam tau {i in Drones} = ceil(d[i] / v / dt);subject to Signal_Propagation {t in Time: t >= max{i in Drones} tau[i]}:    Q[t] = sum{i in Drones} (W[i] * In[i, t - tau[i]] / (d[i]^alpha));```## Пример для 3 дроновПусть:- $v = 3 \times 10^8$ м/с (скорость света)- $d = [100, 200, 300]$ м- $\alpha = 2$Тогда задержки:$$\tau_1 = \frac{100}{3 \times 10^8} = 0.33\ \mu\text{s}$$
